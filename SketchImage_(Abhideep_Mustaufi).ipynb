```python
#Image to Pencil Sketch using python

#Included important libraries

import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Loading and Plotting the original Image
img = cv2.imread("football.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Messi and Ronaldo")
plt.show()
```

```python
#Converting image into gray scale
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
#Inverting the image
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
#Smoothing the image
img_smoothing =cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
#Converting images to pencil sketches
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Sketched Image")
plt.show()
```
